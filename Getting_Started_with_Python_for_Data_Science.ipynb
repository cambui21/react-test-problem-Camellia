###1. Principal Component Analysis (PCA)

Principal Component Analysis (PCA) is a unsupervised method for dimension reduction.

Why PCA and dimension reduction?
- Data with lower dimension are easier to visualize.
- Analyze lower dimension data could be more memory-efficient and computation-efficient.

However, reducing dimension would naturally cause distortion on the data. There is a tradeoff between how much reduction we do and how much distortion we can accept. The key idea behind PCA is how we could reduce dimension of data while minimizing the distortion.

```python
# We can visualize data with two dimensions easily using the following plot
plt.scatter(df['sepal length (cm)'], df['petal length (cm)'], color='red', alpha =1.0)
plt.rcParams['figure.figsize'] = [10,8]
plt.title('Iris Dataset scatter Plot')
plt.xlabel('sepal length (cm)')
plt.ylabel('petal length (cm)')
```

However, it's really hard to visualize data beyond 3D.

```python
# We try to visualize data with four dimensions by using a 3D plot plus the color
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

img = ax.scatter(df['sepal length (cm)'], df['sepal width (cm)'], df['petal length (cm)'] , c=df['petal width (cm)'], cmap=plt.hot())
fig.colorbar(img, label="'petal width (cm)'")

ax.set_xlabel('sepal length (cm)')
ax.set_ylabel('sepal width (cm)')
ax.set_zlabel('petal length (cm)')

ax.set_title('Visualize 4D Data')

plt.show()
```

Let's use PCA to reduce the dimension from 4 to 3.

```python
from sklearn.decomposition import PCA
```

```python
# Create a PCA instance and specify the number of dimension/components we want to keep
pca = PCA(n_components=3)
# We fit our 4 dimensional data
pca.fit(df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']])
# We reduce the dimension of our data 
transformed_data = pca.transform(df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']])
```

Explained variance measures how much variance in original data we captured after transformation. The more varaince we captured, the less distortion there is. We captured 99.48% of variance after reducing the dimension from 4 to 3! This is amazing since we are not losing much variance after reducing the dimension.

```python
print(f"variance explained by each dimension: {pca.explained_variance_ratio_}")
print(f"total variance explained: {sum(pca.explained_variance_ratio_)}")
```

```python
transformed_data[:,0].shape
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

group0 = iris.target == 0
group1 = iris.target == 1
group2 = iris.target == 2

ax.plot(transformed_data[group0,0], transformed_data[group0,1], transformed_data[group0,2], "*", label="setosa")
ax.plot(transformed_data[group1,0], transformed_data[group1,1], transformed_data[group1,2], "o", label="versicolor")
ax.plot(transformed_data[group2,0], transformed_data[group2,1], transformed_data[group2,2], "^", label="virginica")

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title('Iris Dataset After PCA')
plt.legend(loc="upper right")

plt.show()
```

How can we know which type an iris is based on its features?

A simple way of doing such classification is using K-Nearest Neighbors Algorithm. We can use it easily in Python.

```python
from sklearn import neighbors
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
```

We split data into training set and testing set where training set is used to train our model and testing set is used to evaluate our model.

```python
X_train, X_test, y_train, y_test = train_test_split(transformed_data, iris.target, test_size=0.33, random_state=42)
```

```python
# Create the K-Nearest Neighbors classifier instance
clf = neighbors.KNeighborsClassifier(10, weights='uniform')
# Fit our data and the classes of our data
clf.fit(X_train, y_train)
```

```python
# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
x_min, x_max = -5, 5
y_min, y_max = -2, 2
step = 0.1
xx, yy = np.meshgrid(np.arange(x_min, x_max, step),
                        np.arange(y_min, y_max, step))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])
# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

# Plot also the training points
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cmap_bold,
            edgecolor='k', s=20)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title("3-Class classification on Training set")
plt.xlabel('PC1')
plt.ylabel('PC2')

plt.show()
```

```python
# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

# Plot also the training points
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cmap_bold,
            edgecolor='k', s=20)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title("3-Class classification on testing set")
plt.xlabel('PC1')
plt.ylabel('PC2')

plt.show()
```

### 5. Hypothesis Testing and Confidence Interval
We can also do statisical inference in Python. Assume we want to show that the mean of petal length of versicolor and verginica are different:
$$H_0:\mu_0=\mu_1\quad vs\quad H_1:\mu_0\neq\mu_1$$
We can directly use the t-test from the Python scipy library!

```python
from scipy import stats
```

```python
# get the two samples that we want to test
x0 = df['petal length (cm)'].to_numpy()[group1]
x1= df['petal length (cm)'].to_numpy()[group2]

# use t-test on the two samples
stats.ttest_ind(x0, x1, equal_var=False)
```

For significance level of 0.05, we can reject the null hypothesis since the p-value is much smaller!

It's also convinient to calculate **confidence interval** in Python. Let's find the 95% confidence interval for mean of petal length of versicolor.

```python
x0 = df['petal length (cm)'].to_numpy()[group1]
```

```python
print(f"mean: {np.mean(x0)}")
```

```python
# calculate the confidence interval
ci = stats.t.interval(confidence=0.95, df=x0.shape[0]-1, loc=np.mean(x0), scale=stats.sem(x0))
print(f"95% confidence interval: {ci}") 
```

```python
# calculate the confidence interval
ci = stats.t.interval(confidence=0.99, df=x0.shape[0]-1, loc=np.mean(x0), scale=stats.sem(x0))
print(f"99% confidence interval: {ci}") 
```
